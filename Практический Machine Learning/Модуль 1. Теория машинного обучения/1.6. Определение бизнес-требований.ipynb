

Первый этап решения задач МО — определение **бизнес-требований**, то есть понять, что в итоге хочет получить заказчик.

**Выбор цели**. Определите, что именно нужно сделать и действительно ли для достижения этой цели нужно использовать машинное обучение. Не путайте цель и задачи: цель обычно одна, задач несколько. Например, увеличить количество проверяемых гипотез в день химиками на заводе — цель (не до конца определенная). Сделать сервис по распознаванию нарисованных молекул — возможная задача на пути к цели.

**Минимальный жизнеспособный продукт** (MVP, Minimum Viable Product). Не стоит сразу замахиваться на большую цель, подумайте, какое **небольшое решение** задачи может вас удовлетворить. Например, вместо создания полноценного приложения для биометрии можно на первой итерации проекта сделать бота в Телеграме для проверки гипотезы и тестирования модели.

**SMART-цель**. Цель должна быть конкретной, измеримой, достижимой, значимой и ограниченной по времени.

**Выбор метрики**. Определите заранее, как будете **измерять результат**, по каким параметрам. Разные задачи — разные метрики.

**Погружение в предметную область**. Data Science-разработчики — специалисты в алгоритмах МО, а не в терминах систем документооборота. Составьте вместе с заказчиком **словарь** проекта, чтобы говорить на одном языке, проясните специфичные термины и жаргон.
