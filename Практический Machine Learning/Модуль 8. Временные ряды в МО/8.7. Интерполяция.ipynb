Линейная интерполяция

Наиболее частый метод интерполяции — интерполяция линейной функцией P(x)=ax+b.

Соседние известные точки соединяются прямой линией.  Значения неизвестных точек определяются, как значения точек, лежащих на этой прямой.

Когда необходимо приведение к регулярной сетке?

1. Данные получены в произвольные промежутки времени. Гораздо удобнее работать с данными на регулярной сетке.
2. Нужно совместить два временных ряда на регулярных сетках, но эти сетки не совпадают. В этом случае оба ряда приводятся к общей сетке.
