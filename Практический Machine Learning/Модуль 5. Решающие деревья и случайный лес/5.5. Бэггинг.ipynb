Бэггинг

Бэггинг (bagging, bootstrap aggregating) — метод построения композиции алгоритмов, в котором каждый алгоритм строится независимо от других на подвыборках обучающей выборки. Итоговый алгоритм принимает решения посредством голосования всех алгоритмов, т.е. средний ответ, либо усреднение ответов с весами.

Получаемый при агрегации результат — модель, которая сохраняет сдвиг, но может снизить дисперсию. Вернёмся к опросу о кошачьих: не гарантируется, что средний ответ стал ближе к правде, но разброс уменьшился. На практике это позволяет улучшить качество одиночной модели. В самом хорошем случае (при построении) отдельные модели при голосовании должны быть попарно некоррелированными: делать различные ошибки.

Композиция решающих деревьев на практике показывает себя лучше, чем одно дерево. Одиночные деревья зависят от обучающей выборки, затачиваются под неё, и склонны к переобучению. Бэггинг подходит для агрегации решающих деревьев, поскольку композиция алгоритмов при помощи голосования работает наилучшим образом, когда модели различаются.
